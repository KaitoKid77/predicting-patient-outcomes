```python
import pandas as pd
from sklearn.model_selection import train_test_split
from prepare import prep_data
import matplotlib.pyplot as plt
```

```python
df =  pd.read_csv('data/training_v2.csv')
train, test = train_test_split(df, train_size = .80, random_state = 42, stratify = df.hospital_death)
```

```python
train = prep_data(train)
```

```python
train
```

```python
test = prep_data(test)
```

```python
test
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
y = train['hospital_death']
X = train.drop(columns='hospital_death')
```

```python
logit = LogisticRegression()
logit.fit(X, y)
```

```python
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score
```

```python
lr_probs = logit.predict_proba(X)
# keep probabilities for the positive outcome only
lr_probs = lr_probs[:, 1]
```

```python
# calculate scores
# ns_auc = roc_auc_score(testy, ns_probs)
lr_auc = roc_auc_score(y, lr_probs)
# summarize scores
print('Logistic: ROC AUC=%.3f' % (lr_auc))
# calculate roc curves
lr_fpr, lr_tpr, _ = roc_curve(y, lr_probs)
# plot the roc curve for the model
plt.plot(lr_fpr, lr_tpr, marker='.', label='Logistic')
# axis labels
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
# show the legend
plt.legend()
# show the plot
plt.show()
```

```python
y_test = test['hospital_death']
X_test = test.drop(columns='hospital_death')
```

```python
lr_probs = logit.predict_proba(X_test)
# keep probabilities for the positive outcome only
lr_probs = lr_probs[:, 1]
```

```python
# calculate scores
# ns_auc = roc_auc_score(testy, ns_probs)
lr_auc = roc_auc_score(y_test, lr_probs)
# summarize scores
print('Logistic: ROC AUC=%.3f' % (lr_auc))
# calculate roc curves
lr_fpr, lr_tpr, _ = roc_curve(y_test, lr_probs)
# plot the roc curve for the model
plt.plot(lr_fpr, lr_tpr, marker='.', label='Logistic')
# axis labels
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
# show the legend
plt.legend()
# show the plot
plt.show()
```
